```python
import numpy as np
import matplotlib.pyplot as plt # draw graph
# pylot, plot, show, title, xlabel, ylabel, figure, subplot, legend, bar, histogram, scatter
```

```python
# plot
x = np.array([1, 2, 3, 4])
y = np.array([1, 2, 1, 3])
y1 = np.array([2, 1, 2, 5])

plt.plot(x, y)
```

```python
# title, xlabel, ylabel
plt.plot(x, y)
plt.title('My Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# color
plt.plot(x, y, c='r')
plt.title('My Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# 2 graphs in a plot
plt.plot(x, y, x, y1)
plt.title('My Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# draw many plots
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Plot 1')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 2, 2)
plt.plot(x, y1)
plt.title('Plot 2')
plt.xlabel('x')
plt.ylabel('y1')

plt.suptitle('Plots')
```

```python
x1 = ['value 1', 'value 2', 'value 3', 'value 4']
```

```python
# bar, legend /led.nd/
plt.bar(x1, y1)
plt.xlabel('values')
plt.ylabel('y')
plt.title('Bar Graph')
plt.legend(['parameters']) # pass an array
```
